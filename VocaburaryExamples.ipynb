```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import re
import requests
import csv
import MeCab
from tqdm import tqdm_notebook as tqdm
from multiprocessing import Pool
mecab = MeCab.Tagger("-Ochasen")#mecabを呼び出し
f= open("/Users/taichikato/Downloads/jukugo.txt","w+")
```

```python
x = pd.read_csv('~/Downloads/jukugo.csv')[0:2000]['年'].array
```

```python
def generate(data):
    front = ""
    back = ""
    soup = bs(requests.get("http://yourei.jp/" + data).content, 'lxml')
    found = soup.find(id="sentence-1").findAll('span')
    for span in found:
        if data in span.get_text():
            node = mecab.parseToNode(span.get_text())
            rubi = mecab.parseToNode(data).next.feature.split(',')[-2]
            while node:
                split = node.feature.split(',')
                if data in split[-3]:
                    rubi = split[-2]
                    break
                node = node.next
            front = span.get_text().replace(data," [" + rubi + "] ")
            if len(found) > 2:
                back = data + ": " + span.get_text() + "\n 出典: " + found[-2].get_text().replace("より引用","")
            else:
                back = data + ": " + span.get_text() + "\n"
            break
    return front, back
```

```python
for data in tqdm(x):
    front, back = generate(data)
    if front is not "" and back is not "":
        f.write(front + ";\"" + back + "\"\n")
        print(front + ";\"" + back + "\"\n")
f.close()
```
