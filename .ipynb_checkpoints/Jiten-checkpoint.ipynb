```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import re
import requests
```

```python
list = requests.get("https://kanji.jitenon.jp/cat/joyo.html")
```

```python
soup = bs(list.content,'lxml')
```

```python
links = []
for link in soup.findAll('a', attrs={'href': re.compile("^https://kanji.jitenon.jp/kanji")}):
    links.append(link.get('href'))
```

```python
def get_df(res):
    soup = bs(res.content,'lxml')
    table = soup.find_all('table')[1] 
    df = pd.read_html(str(table))
    name = re.compile('(?<=\「).+?(?=\」)').findall(soup.title.string)[0]
    imi = "\n".join(df[0].loc[df[0][0] == '意味'][1].array)
    onyomi = "\n".join(df[0].loc[df[0][0] == '音読み'][1].array)
    kunyomi = "\n".join(df[0].loc[df[0][0] == '訓読み'][1].array)
    final_df = df[0].drop_duplicates(subset=0, keep="first")
    final_df = final_df.set_index(final_df[0])
    final_df.loc["音読み" ,1] = onyomi
    final_df.loc["意味" ,1] = imi
    final_df.loc["訓読み" ,1] = kunyomi
    final_df = final_df.drop(0, axis=1).T
    final_df.insert(0, "漢字", name)
    return final_df
```

```python

link2 = links[200]
res2 = requests.get(link2)
yeet2 = get_df(res2)
```

```python
from tqdm import tqdm_notebook as tqdm
```

```python
data = pd.DataFrame()
for link in tqdm(links):
    row = get_df(requests.get(link))
    data = data.append(row)
```

```python
data['裏'] = data['訓読み'] + "\n" + data['音読み'] + "\n\n意味：" + data['意味']
data[['漢字','裏']].to_csv(r'~/Downloads/clean.txt', index = None, header=True, sep=";")
```
